# MVP

```python
from random import randint
```

At first, I was planning to search for a function that would allow me to look for the exact user's input in the 'title' column of the dataframe. I then came to that page : https://stackoverflow.com/questions/43758249/exact-match-string-in-panda-column

The offered solution to this other problem inspired me to make a subdataframe from the popvortex_100 based on the title/user's input and check if this subset has rows or not.

```python
def mvp():
    song_input = input('Please fill in the title of a song you like :')
    top_100_check = popvortex_100[popvortex_100['title'] == song_input]
    if len(top_100_check['title']) == 0 :
        return 'Sorry, this song is not part of the PopVortex Top 100 at the moment.'
    else :
        for i in range(4):   # I want to avod here to suggest the same song the user just filled in.
            # I plan 4 loops, but the probability the random choice ends three times on the initial user input is almost zero.
            index = randint(1,100)
            proposed_title = popvortex_100['title'][index]
            if proposed_title != song_input:
                break
        proposed_artist = popvortex_100['artist'][index]
        output = 'Great choice ! This song in in the PopVortex Top 100. You should listen to ' + proposed_title + ' by ' + proposed_artist + '. You might like it !'
        return output
```

It works for one list of songs. Let's make a function to scrap several lists.

```python
def list_scraping(genre, year):
    # create URL
    url_scrap = "https://playback.fm/charts/"+ genre +"/"+ str(year)
    # Create request
    resp_scrap = requests.get(url_scrap)
    # Parse the response
    soup_scrap = BeautifulSoup(resp_scrap.content, "html.parser")
    
    # Get list of titles and artists
    title_scrap = []
    artist_scrap = []
    for song in soup_scrap.select("span.song > a"):
        title_scrap.append(song.get_text().replace('\n',''))
    for artist in soup_scrap.select("a.artist"):
        artist_scrap.append(artist.get_text().replace('\n',''))
    
    # create dataframe
    data = pd.DataFrame({"title":title_scrap, "artist":artist_scrap })
    
    return data
```

I will create a dictionnary of dataframes to store the genre, year and matching dataframe.

```python
genre_list = ['rock', 'country', 'rnb', 'brasil']
year_range = range(2000,2022)

song_list_dict = {}

for genre in genre_list:
    for year in year_range:
        song_list_dict[genre+'_'+str(year)+'_list'] = list_scraping(genre, year)

song_list_dict
```

```python
# How to check the available years
url_genre = "https://playback.fm/charts/brasil"
resp_genre = requests.get(url_genre)
soup_genre = BeautifulSoup(resp_genre.content, "html.parser")
```

```python
def list_scraping_2(genre, year):
    # create URL
    url_genre = "https://playback.fm/charts/"+ genre
    url_genre_year = "https://playback.fm/charts/"+ genre +"/"+ str(year)
    # Create request
    resp_genre = requests.get(url_genre)
    resp_genre_year = requests.get(url_genre_year)
    # Parse the responses
    soup_genre = BeautifulSoup(resp_genre.content, "html.parser")
    soup_genre_year = BeautifulSoup(resp_genre_year.content, "html.parser")
    
    year_range_genre = []
    for j in soup_genre.select("div.allYears > a"):
        year_range_genre.append(j.get_text().replace('\n','')[0: 4])
    
    if str(year) not in year_range_genre:
        return 'year not available for this genre'
    else:
        # Get list of titles and artists
        title_scrap = []
        artist_scrap = []
        for song in soup_genre_year.select("span.song > a"):
            title_scrap.append(song.get_text().replace('\n',''))
        for artist in soup_genre_year.select("a.artist"):
            artist_scrap.append(artist.get_text().replace('\n',''))

        # create dataframe
        data = pd.DataFrame({"title":title_scrap, "artist":artist_scrap })

        return data
```

```python
song_list_dict_2 = {}

for genre in genre_list:
    for year in year_range:
        if list_scraping_2(genre, year) != 'year not available for this genre':
            song_list_dict_2[genre+'_'+str(year)+'_list'] = list_scraping_2(genre, year)

song_list_dict_2
```

This way does not work, as the logical test in the if statement is ambiguous for non empty dataframes...
But I thath helped me think about an even simpler way : use back the original version of my funtcion and test the length of the returned dataframe. If there is no row, we do not include it in the dictionnary.

```python
song_list_dict_2 = {}

for genre in genre_list:
    for year in year_range:
        if len(list_scraping(genre, year)['title']) > 0:
            song_list_dict_2[genre+'_'+str(year)+'_list'] = list_scraping(genre, year)

song_list_dict_2
```
